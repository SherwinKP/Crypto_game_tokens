![alt text](https://lvena26b6e621o8sl2qkx1ql-wpengine.netdna-ssl.com/wp-content/uploads/2018/02/XPecosystem.png "Logo Title Text 1")

- Fork off PeerCoin, a crypto aiming to achieve energy efficiency
- Bitcoin's security relies on miners. As time goes on, Bitcoin will be forced to move transactions "off block-chain". 
- If Bitcoin moves off block-chain, miners cannot be compensated for these transactions.
- If miners cannot be compensated, they won't mine. 
- If miners don't mine, network security goes down as time goes on. As time goes on, Bitcoin may become increasingly less secure and more risky to use as a safe haven or a way to save and store very large sums of money.
- PeerCoin uses proof-of-stake for minting and tx processing, where stake is to the use of currency itself (ownership) to achieve certain goals. 

Proof of stake example https://github.com/ethereum/casper/blob/master/casper/contracts/simple_casper.v.py

1. The validators stake a portion of their Ethers as stake.
2. After that, they will start validating the blocks. Meaning, when they discover a block which they think can be added to the chain, they will validate it by placing a bet on it.
3. If the block gets appended, then the validators will get a reward proportionate to their bets.
4. However, if a validator acts in a malicious manner and tries to do a “nothing at stake”, they will immediately be reprimanded, and all of their stake is going to get slashed.

## Chimaera

![alt text](http://todaycoinindex.com/wp-content/uploads/2018/02/1_8SZijen9WK9609HIqPRKnA.jpg "Logo Title Text 1")

- Chimaera is set to provide developers with tools to create and launch games at a lower cost
- Video games running on it are powered by CHI tokens on its blockchain. - - It will have an API so that games written in any language can connect to it, giving designers the opportunity to design games in their language of choice. 
- incorporates Game Channels to add-in near limitless scalability that can accommodate countless games and players. 
- Chimaera offers an off-chain technology that let gamers only process and store states for games they are participating in. This is unlike Ethereum in which every full node fully processes and stores the state for every contract.
- compatible with popular game engines like Unity and Unreal Engine. 
- All trading transactions are safe through Atomic Transactions. Both parties must sign the transaction or it becomes invalidated after a pre-determined time-out period.
